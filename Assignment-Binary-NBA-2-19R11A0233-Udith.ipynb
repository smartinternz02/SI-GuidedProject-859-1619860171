```python
import pandas as pd
import numpy as np
```

```python
d=pd.read_csv(r"nba_logreg.csv")
```

```python
d.head()
```

```python
d.isnull().sum()
```

```python
d["3p"].fillna(d["3p"].mean(),inplace = True)
```

```python
x=d.iloc[:,1:20].values
y=d.iloc[:,20].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test  = train_test_split(x,y , test_size = 0.2 , random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc =StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense 
```

```python
ann=Sequential()
```

```python
ann.add(Dense(19,kernel_initializer = "random_uniform", activation ="relu"))
```

```python
ann.add(Dense(38,kernel_initializer = "random_uniform", activation ="relu"))
```

```python
ann.add(Dense(1,kernel_initializer = "random_uniform", activation ="sigmoid"))
```

```python
ann.compile("adam","binary_crossentropy",metrics=["accuracy"])
```

```python
ann.fit(x_train,y_train,batch_size=32,epochs=500)
```

```python
ypred=ann.predict(x_test)
```
